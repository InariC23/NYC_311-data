# **What are the most frequent 311 complaints in NYC?**

```python
%%bigquery --project capstone-project-347718

WITH request_query 
AS
(SELECT
 created_date,
 complaint_type,
 agency,
 borough,
 EXTRACT(isoyear from created_date) as year
FROM `bigquery-public-data.new_york_311.311_service_requests`)

SELECT
  complaint_type,
  COUNT(complaint_type) AS `complaint_num` 
FROM request_query
WHERE year >= 2019 
AND year <= 2021
GROUP BY 
complaint_type
ORDER BY 
  `complaint_num` DESC 
LIMIT 10
```

# **What are the least frequent 311 complaints in NYC?**


```python
%%bigquery --project capstone-project-347718

WITH requests_query 
AS
(SELECT
 created_date,
 complaint_type,
 agency,
 borough,
 EXTRACT(isoyear from created_date) as year
FROM `bigquery-public-data.new_york_311.311_service_requests`)

SELECT
  complaint_type,
  borough,
  COUNT(complaint_type) AS `complaint_num` 
FROM requests_query
WHERE year >= 2019 
AND year <= 2021
GROUP BY 
complaint_type,borough
HAVING `complaint_num` > 10
ORDER BY 
  `complaint_num` ASC 
  LIMIT 15 
```

# **On average how many hours does it take for a agency to close a case?**

```python
%%bigquery --project capstone-project-347718 

WITH hour_query 
AS
(SELECT
 created_date,
 complaint_type,
 agency,
 borough,
 status,
 EXTRACT(isoyear from created_date) as year,
 TIMESTAMP_DIFF(closed_date, created_date, HOUR) as hours
FROM `bigquery-public-data.new_york_311.311_service_requests`)

SELECT
  complaint_type,
  agency,
  status,
  borough,
  avg(hours) as hours,
  COUNT(complaint_type) AS `value_occurrence` 
FROM hour_query
WHERE year >= 2019 
AND year <= 2021 AND hours IS NOT NULL
GROUP BY agency,status,borough,complaint_type
ORDER BY 
  `value_occurrence` DESC 
```
